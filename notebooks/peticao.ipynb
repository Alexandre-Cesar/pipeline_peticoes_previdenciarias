```python
modelo = Document(f"{path}template/Peticao_MODELO_Auxilio_Doenca_Maria da Silva_2026-06-22.docx")

novo = Document()

novo.styles.element[:] = modelo.styles.element[:]
```

```python
for linha in peticao.split('\n'):
    linha = linha.strip()

    if linha:
        p = novo.add_paragraph(linha)
        p.paragraph_format.space_after = Pt(0)
    else:
        # Linha em branco
        novo.add_paragraph()

novo.save(f"{path}output/peticao_joao_da_silva.docx")
```
