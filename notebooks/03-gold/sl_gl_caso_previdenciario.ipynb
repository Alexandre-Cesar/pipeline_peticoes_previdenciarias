```python
from pyspark.sql import functions as f

df_rg = spark.read.table("silver.rg").select("nome", "cpf", "data_nascimento")
df_residencia = spark.read.table("silver.residencia").select("cpf", "endereco")
df_indef = spark.read.table("silver.indeferimento").select("cpf","numero_beneficio", "status", "motivo")
df_laudo = spark.read.table("silver.laudo").select("nome", "data_nascimento", "profissao", "conclusao", "cid10", "descricao_cid", "crm", "especialidade", "possui_sequela_permanente", "possui_reducao_capacidade")
```

```python
df = df_rg.join(df_residencia, on='cpf', how='inner').join(df_indef, on='cpf', how='inner').join(df_laudo, on=['nome', 'data_nascimento'],how='inner')
```

```python
df.write.mode("overwrite").saveAsTable("gold.tb_caso_previdenciario")
```
