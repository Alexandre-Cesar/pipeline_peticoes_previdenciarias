```python
from pyspark.sql import functions as f
df= spark.read.table("bronze.indeferimento")
```

```python
df = df.withColumn(
    "cpf",
    f.regexp_replace("cpf", r"\D", "")
)
#para indeferimento
df = df.withColumn(
    "numero_beneficio",
    f.regexp_replace("numero_beneficio", r"\D", "")
)
```

```python
df.write.mode("overwrite").saveAsTable("silver.indeferimento")
```
