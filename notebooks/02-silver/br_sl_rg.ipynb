```python
from pyspark.sql import functions as f

df = spark.read.table("bronze.rg")
```

```python
df = df.withColumn(
    "data_nascimento",
    f.to_date("data_nascimento", "dd/MM/yyyy")
)
```

```python
df = df.withColumn(
    "cpf",
    f.regexp_replace("cpf", r"\D", "")
)
```

```python
df.write.mode("overwrite").saveAsTable("silver.rg")
```
