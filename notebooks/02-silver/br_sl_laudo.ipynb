```python
from pyspark.sql import functions as f

df = spark.read.table("bronze.laudo")
```

```python
df = df.withColumn(
    "data_nascimento",
    f.to_date("data_nascimento", "dd/MM/yyyy")
)
```

```python
df = (
    df
    .withColumn(
        "possui_sequela_permanente",
        f.when(
            f.upper(f.col("conclusao")).contains("SEQUELA PERMANENTE"),
            True
        ).otherwise(False)
    )
    .withColumn(
        "possui_reducao_capacidade",
        f.when(
            f.upper(f.col("conclusao")).contains("REDUÇÃO")
            | f.upper(f.col("conclusao")).contains("REDUCAO"),
            True
        ).otherwise(False)
    )
)
```

```python
df.write.mode("overwrite").saveAsTable("silver.laudo")
```
